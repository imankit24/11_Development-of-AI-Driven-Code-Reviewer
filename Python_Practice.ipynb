## 1. NumPy Practice
Basic array creation and operations.

```python

import numpy as np

arr = np.array([10, 20, 30, 40])
print("Array:", arr)
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
```

## 2. Pandas Practice
Working with tabular data using DataFrames.

```python

import pandas as pd

data = {
    "Name": ["Ankit Raj Jha", "Dev Agrawal", "Harsh Parvani"],
    "Score": [85, 90, 88]
}

df = pd.DataFrame(data)
df
```

## 3. Matplotlib Practice
Basic data visualization.

```python

import matplotlib.pyplot as plt

scores = df["Score"]
plt.plot(scores)
plt.title("Student Scores")
plt.xlabel("Index")
plt.ylabel("Score")
plt.show()
```

## 4. AST Module Practice
Simple code structure inspection.

```python

import ast

code = "x = 5\nprint(x)"
tree = ast.parse(code)

for node in ast.walk(tree):
    if isinstance(node, ast.Assign):
        print("Assignment statement found")
```
